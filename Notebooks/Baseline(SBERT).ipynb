# SBERT

```python
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
```

```python
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
amazon_texts = df_amazon['review'] 
amazon_labels = df_amazon['sentiment']

imdb_texts = df_imdb['review']
imdb_labels = df_imdb['sentiment']


amazon_embeddings = sbert_model.encode(amazon_texts, convert_to_numpy=True)

imdb_embeddings = sbert_model.encode(imdb_texts, convert_to_numpy=True)

```

```python

X_train, X_val, y_train, y_val = train_test_split(amazon_embeddings, amazon_labels, test_size=0.1, random_state=42)
```

```python
clf = LogisticRegression(max_iter=2000)
clf.fit(X_train, y_train)
```

```python
imdb_preds = clf.predict(imdb_embeddings)
imdb_acc = accuracy_score(imdb_labels, imdb_preds)
```

```python
report_imdb = classification_report(imdb_labels, imdb_preds, output_dict=True)
```

```python
report_log_df = pd.DataFrame(report_imdb).transpose()

report_log_df
```
